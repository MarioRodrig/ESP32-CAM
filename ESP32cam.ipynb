# Documentation for Utilizing the ESP32-CAM

--------------------------------------------

### A concise summary of my approach to leveraging the capabilities of the ESP32-CAM

## Step 1: Necessary items

- Esp32-CAM

- You'll need five male to male wires to connect the flash device to the Esp32-CAM

- CP2102 flash device 

## Step 2: Wiring Up 

Flash Device wiring:

- on the flash device we plug in 5v and the ground

- Then the RXD ( orange ) and TXD ( blue )

Esp32-CAM wiring:

- Plug-in the 5v wire to 5v on top left of Esp32 and below it plug-in ground.  

- Here we plug in our TXD ( Blue ) to UnR pin, and then the RXD ( Orange ) to UoT pin.

- Last wire is used to jump GND to I00 pin (refer to the first image to see where they're located)
